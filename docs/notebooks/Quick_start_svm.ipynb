```python
from fromscratchtoml.models import svm
import torch as ch
import numpy as np

from fromscratchtoml.toolbox import binary_visualize
from fromscratchtoml.toolbox.random import Distribution

%matplotlib inline
```

## Linearly seperable data

```python
X1 = Distribution.linear(pts=10,
               mean=[8, 10],
               covr=[[1.5, 1], [1, 1.5]])
X2 = Distribution.linear(pts=10,
               mean=[9, 5],
               covr=[[1.5, 1], [1, 1.5]])

Y1 = ch.ones(X1.size()[0])
Y2 = -ch.ones(X2.size()[0])

X = ch.cat((X1, X2))
y = ch.cat((Y1, Y2))
```

```python
clf_lin = svm.SVC(kernel='linear')
clf_lin.fit(X, y)
```

```python
X1 = Distribution.linear(pts=100,
               mean=[8, 10],
               covr=[[1.5, 1], [1, 1.5]])
X2 = Distribution.linear(pts=100,
               mean=[9, 5],
               covr=[[1.5, 1], [1, 1.5]])
X = ch.cat((X1, X2))
_, projections = clf_lin.predict(X, return_projection=True)
```

```python
binary_visualize(X, clf=clf_lin)
```

## RBF

```python
X1 = Distribution.radial_binary(pts=10,
               mean=[0, 0],
               start=1,
               end=2)
X2 = Distribution.radial_binary(pts=10,
               mean=[0, 0],
               start=4,
               end=5)

Y1 = ch.ones(X1.size()[0])
Y2 = -ch.ones(X1.size()[0])

X = ch.cat((X1, X2))
y = ch.cat((Y1, Y2))
```

```python
clf = svm.SVC(kernel='rbf', gamma=0.2)
clf.fit(X, y)

binary_visualize(X, clf=clf, ylim=[-5, 5])
```

```python
clf = svm.SVC(kernel='rbf', gamma=0.4)
clf.fit(X, y)

binary_visualize(X, clf=clf)
```

```python
clf = svm.SVC(kernel='rbf', gamma=1)
clf.fit(X, y)

binary_visualize(X, clf=clf, ylim=[-4, 5])
```

```python
# default is set to kernel='rbf', gamma=2
clf = svm.SVC()
clf.fit(X, y)

binary_visualize(X, clf=clf, ylim=[-4, 5])
```

```python
clf = svm.SVC(kernel='rbf', gamma=10)
clf.fit(X, y)

binary_visualize(X, clf=clf, ylim=[-4, 5])
```

```python
X1 = Distribution.linear(pts=100,
               mean=[6, 10],
               covr=[[1.5, 1], [1, 1.5]])
X2 = Distribution.linear(pts=100,
               mean=[9, 5],
               covr=[[1.5, 1], [1, 1.5]])
X3 = Distribution.linear(pts=100,
               mean=[-9, -5],
               covr=[[1.5, 1], [1, 1.5]])
X4 = Distribution.linear(pts=100,
               mean=[-6, -10],
               covr=[[1.5, 1], [1, 1.5]])

Y1 = -1*ch.ones(X1.size()[0])
Y2 = 1*ch.ones(X2.size()[0])
Y3 = 2*ch.ones(X3.size()[0])
Y4 = 3*ch.ones(X4.size()[0])

X = ch.cat((X1, X2, X3 ,X4))
y = ch.cat((Y1, Y2, Y3, Y4))
```

```python
CLF = svm.SVC()
classifiers=CLF.fit(X, y) 
```

```python
for clf in classifiers:
    binary_visualize(X, clf=clf)
```

```python
%% cache
# Multivariate SVC
uniques, ind = np.unique(y.numpy(), return_index=True)
print(len(uniques))
y_list = [np.where(y.numpy() == u, 1, -1) for u in uniques]

classifiers = []
for y_i in y_list: 
    clf = svm.SVC(kernel='linear', C=1000) 
    clf.fit(X, y_i) 
    classifiers.append(clf)

predictions = []
for x in ch.Tensor(X):
    for i, clas in enumerate(classifiers):
        if classifiers[i].predict(x)[0] == 1:
            predictions.append(y[ind[i]])
            break
        if i==len(classifiers)-1:
            predictions.append(0)
```
